```python
import pandas as pd

ranaumar = pd.DataFrame({

    "Student_ID": [
        "SU74-BSSEM-F23-018",
        "SU74-BSSEM-F23-019",
        "SU74-BSSEM-F23-020",
        "SU74-BSSEM-F23-021",
        "SU74-BSSEM-F23-022"
    ],

    "Name": [
        "Rana Umar",
        "Ali",
        "Ahmed",
        "Hassan",
        "Usman"
    ],

    "Semester":[6,6,6,5,7],

    "Python":[90,80,75,88,92],

    "Machine Learning":[88,75,70,85,91],

    "DBMS":[89,72,74,80,90],

    "CGPA":[3.75,3.20,3.10,3.45,3.82],

    "Internship Status":[
        "No",
        "Yes",
        "No",
        "Yes",
        "No"
    ]
})

print("Complete DataFrame")
print(ranaumar)

print("\nNames")
print(ranaumar["Name"])

print("\nName and CGPA")
print(ranaumar[["Name","CGPA"]])

print("\nFirst Student")
print(ranaumar.iloc[0])

print("\nThird Student")
print(ranaumar.iloc[2])

print("\nFirst Three Students")
print(ranaumar.iloc[0:3])

print("\nSpecific Value")
print(ranaumar.iloc[1,3])



print("Python > 85")
print(ranaumar[ranaumar["Python"] > 85])
print("\nCGPA > 3.5")
print(ranaumar[ranaumar["CGPA"] > 3.5])


print("\nInternship Status = No")
print(ranaumar[ranaumar["Internship Status"]=="No"])

print("\nQuery Example")

result = ranaumar.query("CGPA > 3.3 and Python > 85")

print(result)
print("\nWhere Example")
print(ranaumar.where(ranaumar["Python"]>80))
```
