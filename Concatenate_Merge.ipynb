```python
import pandas as pd

students = pd.DataFrame({

    "Student_ID":[
        "SU74-BSSEM-F23-018",
        "SU74-BSSEM-F23-019",
        "SU74-BSSEM-F23-020"
    ],

    "Name":[
        "Rana Umar",
        "Aqsa",
        "Aqsa"
    ],

    "Department":[
        "Software Engineering",
        "Software Engineering",
        "Computer Science"
    ]

})

results = pd.DataFrame({

    "Student_ID":[
        "SU74-BSSEM-F23-018",
        "SU74-BSSEM-F23-019",
        "SU74-BSSEM-F23-020"
    ],

    "Python":[90,80,75],

    "Machine Learning":[88,78,70],

    "CGPA":[3.75,3.20,3.10]

})

print("Students Data")
print(students)

print("\nResults Data")
print(results)

# Merge

merged = pd.merge(students,results,on="Student_ID")

print("\nMerged DataFrame")

print(merged)

internship = pd.DataFrame({

    "Student_ID":[
        "SU74-BSSEM-F23-018",
        "SU74-BSSEM-F23-020"
    ],

    "Internship":[
        "No",
        "Yes"
    ]

})


print("\nLeft Join")

print(pd.merge(students,internship,on="Student_ID",how="left"))


print("\nRight Join")

print(pd.merge(students,internship,on="Student_ID",how="right"))


print("\nInner Join")

print(pd.merge(students,internship,on="Student_ID",how="inner"))


print("\nOuter Join")

print(pd.merge(students,internship,on="Student_ID",how="outer"))


semester6 = pd.DataFrame({

    "Name":[
        "Rana Umar",
        "Fatima"
    ],

    "Semester":[6,6]

})

semester7 = pd.DataFrame({

    "Name":[
        "Jimmy",
        "Ishal"
    ],

    "Semester":[7,7]

})

combined = pd.concat([semester6,semester7])

print("\nConcatenate")
print(combined)
```
