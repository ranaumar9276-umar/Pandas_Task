```python
import pandas as pd

ranaumar = pd.DataFrame({

    "Name":[
        "Rana Umar",
        "Aqsa",
        "Jimmy",
        "Fatima",
        "Ishal"
    ],

    "Assignment_Date":[
        "2026-06-10",
        "2026-06-12",
        "2026-06-14",
        "2026-06-18",
        "2026-06-20"
    ],

    "Submission_Date":[
        "2026-06-15",
        "2026-06-17",
        "2026-06-19",
        "2026-06-22",
        "2026-06-25"
    ]

})

print(ranaumar)

# Convert String into Date

ranaumar["Assignment_Date"]=pd.to_datetime(ranaumar["Assignment_Date"])

ranaumar["Submission_Date"]=pd.to_datetime(ranaumar["Submission_Date"])

print("\nData Types")

print(ranaumar.dtypes)

# Extract Year

ranaumar["Year"]=ranaumar["Submission_Date"].dt.year

# Extract Month

ranaumar["Month"]=ranaumar["Submission_Date"].dt.month

# Extract Day

ranaumar["Day"]=ranaumar["Submission_Date"].dt.day

# Extract Day Name

ranaumar["Day_Name"]=ranaumar["Submission_Date"].dt.day_name()

print("\nUpdated DataFrame")

print(ranaumar)
```
