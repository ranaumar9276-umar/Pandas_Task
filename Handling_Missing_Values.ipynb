```python
import pandas as pd
import numpy as np

ranaumar = pd.DataFrame({

    "Name":[
        "Rana Umar",
        "Aqsa",
        "Jimmy",
        "Fatima",
        "Ishal"
    ],

    "Python":[
        90,
        np.nan,
        75,
        88,
        np.nan
    ],

    "Machine Learning":[
        88,
        80,
        np.nan,
        85,
        91
    ],

    "CGPA":[
        3.75,
        3.20,
        np.nan,
        3.45,
        3.82
    ]

})

print("Original Data")

print(ranaumar)

#  Missing Values

print("\nMissing Values")

print(ranaumar.isnull())

# Count Missing Values

print("\nTotal Missing")

print(ranaumar.isnull().sum())

# Fill Missing Values

ranaumar["Python"]=ranaumar["Python"].fillna(0)

ranaumar["Machine Learning"]=ranaumar["Machine Learning"].fillna(80)

ranaumar["CGPA"]=ranaumar["CGPA"].fillna(ranaumar["CGPA"].mean())

print("\nAfter Filling Missing Values")

print(ranaumar)

# Drop Missing Values

print("\nDrop Missing")

print(ranaumar.dropna())
```
