```python
import pandas as pd

ranaumar = pd.DataFrame({

    "Department":[
        "Software Engineering",
        "Software Engineering",
        "Computer Science",
        "Computer Science",
        "Software Engineering"
    ],

    "Name":[
        "Rana Umar",
        "Aqsa",
        "Jimmy",
        "Fatima",
        "Ishal"
    ],

    "Python":[
        90,
        80,
        75,
        88,
        92
    ],

    "Machine Learning":[
        88,
        78,
        70,
        85,
        91
    ],

    "CGPA":[
        3.75,
        3.20,
        3.10,
        3.45,
        3.82
    ]

})

print("Complete Data")

print(ranaumar)

# Aggregation

print("\nAverage Python Marks")

print(ranaumar["Python"].mean())

print("\nMaximum CGPA")

print(ranaumar["CGPA"].max())

print("\nMinimum CGPA")

print(ranaumar["CGPA"].min())

print("\nTotal Python Marks")

print(ranaumar["Python"].sum())

# Group By Department

print("\nGroup By Department")

print(ranaumar.groupby("Department").mean(numeric_only=True))

# Group By Department and Count

print("\nStudents Count")

print(ranaumar.groupby("Department").count())

# Group By Department and Maximum CGPA

print("\nHighest CGPA")

print(ranaumar.groupby("Department")["CGPA"].max())
```
